```python

import math
# Step 1: Mean and standard deviation for standardizing the data
def mean(values):
    return sum(values) / len(values)
```

```python


# Step 2: Standardize the dataset
def standard_deviation(values, mean_value):
    return math.sqrt(sum((x - mean_value) ** 2 for x in values) / len(values))

def standardize_data(matrix):
    mean_values = [mean(col) for col in zip(*matrix)]  # Mean of each column
    std_devs = [standard_deviation(col, mean_values[i]) for i, col in enumerate(zip(*matrix))]  # Std dev of each column
    
    # Standardize (z-score normalization)
    standardized_matrix = []
    for row in matrix:
        standardized_row = [(row[i] - mean_values[i]) / std_devs[i] for i in range(len(row))]
        standardized_matrix.append(standardized_row)
    
    return standardized_matrix
```

```python
# Step 3: Calculate covariance matrix
def covariance_matrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    
    # Initialize covariance matrix
    cov_matrix = [[0 for _ in range(m)] for _ in range(m)]
    
    for i in range(m):
        for j in range(m):
            cov_ij = sum((matrix[k][i] * matrix[k][j]) for k in range(n)) / (n - 1)
            cov_matrix[i][j] = cov_ij
            
    return cov_matrix
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Standardize the data
def standardize_data(matrix):
    mean = np.mean(matrix, axis=0)
    std_dev = np.std(matrix, axis=0)
    standardized_matrix = (matrix - mean) / std_dev
    return standardized_matrix

# Compute the covariance matrix
def covariance_matrix(matrix):
    return np.cov(matrix, rowvar=False)

# Subtract lambda times the identity matrix from the given matrix
def subtract_lambda_identity(matrix, lambd):
    return matrix - lambd * np.identity(len(matrix))

# Calculate the determinant of a matrix
def determinant(matrix):
    return np.linalg.det(matrix)

# Find eigenvalues
def find_eigenvalues(matrix):
    eigenvalues = []
    n = len(matrix)
    
    for guess in range(-100, 101):  # Guess eigenvalues from -100 to 100
        mat_minus_lambda = subtract_lambda_identity(matrix, guess)
        if np.isclose(determinant(mat_minus_lambda), 0):
            eigenvalues.append(guess)
    
    return list(set(eigenvalues))  # Return unique eigenvalues

# Find eigenvectors
def find_eigenvectors(matrix, eigenvalue):
    eigvals, eigvecs = np.linalg.eig(matrix)
    idx = np.where(np.isclose(eigvals, eigenvalue))[0]
    return eigvecs[:, idx].flatten()

# Project data onto top-k eigenvectors
def project_data(matrix, eigenvectors):
    projected_data = np.dot(matrix, np.array(eigenvectors).T)
    return projected_data

# Main PCA function
def pca(matrix, num_components):
    # Step 1: Standardize data
    standardized_matrix = standardize_data(matrix)
    
    # Step 2: Compute covariance matrix
    cov_matrix = covariance_matrix(standardized_matrix)
    
    # Step 3: Find eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # Step 4: Sort eigenvectors by eigenvalues (descending order)
    eigenvalue_eigenvector_pairs = sorted(zip(eigenvalues, eigenvectors.T), key=lambda x: x[0], reverse=True)
    
    # Step 5: Select top-k eigenvectors
    top_eigenvectors = [pair[1] for pair in eigenvalue_eigenvector_pairs[:num_components]]
    
    # Step 6: Project data onto the new space
    projected_data = project_data(standardized_matrix, top_eigenvectors)
    
    return projected_data

# Generate circular data
theta = np.linspace(0, 2 * np.pi, 100)
r = 5
x = r * np.cos(theta)
y = r * np.sin(theta)

# Combine into a dataset
circular_data = np.column_stack((x, y))

# Plot the circular data
plt.scatter(circular_data[:, 0], circular_data[:, 1])
plt.title("Circular Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# Perform PCA to reduce to 1 principal component
num_components = 1
projected_data = pca(circular_data, num_components)

# Output the projected data
print("Projected Data:")
for row in projected_data:
    print(row)
```

```python
import math

# Helper function: Transpose of a matrix
def transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

# Helper function: Matrix multiplication
def matrix_multiply(A, B):
    result = [[0] * len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            result[i][j] = sum(A[i][k] * B[k][j] for k in range(len(B)))
    return result

# Helper function: Subtract lambda * Identity matrix from A
def subtract_lambda_identity(matrix, lambd):
    size = len(matrix)
    return [[matrix[i][j] - (lambd if i == j else 0) for j in range(size)] for i in range(size)]

# Helper function: Normalize a vector
def normalize(vector):
    magnitude = math.sqrt(sum(x ** 2 for x in vector))
    return [x / magnitude for x in vector]

# QR Decomposition: Efficient way to find eigenvalues and eigenvectors
def qr_decomposition(matrix):
    # Using Gram-Schmidt process for QR decomposition
    def gram_schmidt(matrix):
        Q = []
        for v in matrix:
            v = v[:]
            for q in Q:
                dot_product = sum(v[i] * q[i] for i in range(len(v)))
                v = [v[i] - dot_product * q[i] for i in range(len(v))]
            norm = math.sqrt(sum(x ** 2 for x in v))
            Q.append([x / norm for x in v])
        return Q

    A = matrix[:]
    for i in range(len(A)):
        for j in range(i + 1, len(A)):
            A[j] = [A[j][k] - A[i][k] * (A[j][i] / A[i][i]) for k in range(len(A))]

    Q = gram_schmidt(transpose(A))
    R = matrix_multiply(transpose(Q), matrix)
    
    return Q, R

# Power iteration method for finding eigenvalues and eigenvectors
def power_iteration(matrix, num_simulations=1000):
    b_k = [1 for _ in range(len(matrix))]
    for _ in range(num_simulations):
        # Multiply the matrix by the vector
        b_k1 = [sum(matrix[i][j] * b_k[j] for j in range(len(matrix))) for i in range(len(matrix))]
        # Normalize the resulting vector
        b_k = normalize(b_k1)
    
    # Eigenvalue: Approximation as dot product of Av and v
    Av = [sum(matrix[i][j] * b_k[j] for j in range(len(matrix))) for i in range(len(matrix))]
    eigenvalue = sum(b_k[i] * Av[i] for i in range(len(matrix)))
    
    return eigenvalue, b_k

# Function to find all eigenvalues and eigenvectors
def find_eigenvalues_eigenvectors(matrix):
    eigenvalues = []
    eigenvectors = []
    
    Q, R = qr_decomposition(matrix)
    A_k = matrix_multiply(R, Q)

    for _ in range(len(matrix)):
        eigenvalue, eigenvector = power_iteration(A_k)
        eigenvalues.append(eigenvalue)
        eigenvectors.append(eigenvector)
        
        A_k = subtract_lambda_identity(A_k, eigenvalue)
    
    return eigenvalues, eigenvectors

# Function to compute covariance matrix
def covariance_matrix(matrix):
    n = len(matrix)
    m = len(matrix[0])
    mean_vector = [sum(row[i] for row in matrix) / n for i in range(m)]
    
    cov_matrix = [[0 for _ in range(m)] for _ in range(m)]
    for i in range(m):
        for j in range(m):
            cov_matrix[i][j] = sum((matrix[k][i] - mean_vector[i]) * (matrix[k][j] - mean_vector[j]) for k in range(n)) / (n - 1)
    
    return cov_matrix

# Function to perform PCA
def pca(matrix, num_components):
    # Step 1: Compute covariance matrix
    cov_matrix = covariance_matrix(matrix)
    
    # Step 2: Find eigenvalues and eigenvectors of covariance matrix
    eigenvalues, eigenvectors = find_eigenvalues_eigenvectors(cov_matrix)
    
    # Step 3: Sort eigenvectors by descending eigenvalues
    eigenvalue_eigenvector_pairs = sorted(zip(eigenvalues, eigenvectors), key=lambda x: x[0], reverse=True)
    
    # Step 4: Select the top-k eigenvectors
    top_eigenvectors = [pair[1] for pair in eigenvalue_eigenvector_pairs[:num_components]]
    
    # Step 5: Project data onto the new k-dimensional subspace
    projected_data = matrix_multiply(matrix, transpose(top_eigenvectors))
    
    return projected_data

# Example usage:
data = [
    [2.5, 2.4],
    [0.5, 0.7],
    [2.2, 2.9],
    [1.9, 2.2],
    [3.1, 3.0],
    [2.3, 2.7],
    [2.0, 1.6],
    [1.0, 1.1],
    [1.5, 1.6],
    [1.1, 0.9]
]

# Perform PCA to reduce to 1 principal component
num_components = 1
projected_data = pca(np.array(X), num_components)

# Output the projected data
print("Projected Data:")
for row in projected_data:
    print(row)
```
